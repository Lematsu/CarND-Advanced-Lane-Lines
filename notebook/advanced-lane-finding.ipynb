## Advanced Lane Finding


#### Goals
The goal is to write a software pipeline to identify the lane boundaries in a video.

The goals / steps of this project are the following:
* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transform, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

```python
# Plot image
def plot_image(img, title,color=None):
    '''
    plot given image with title and color
    '''
    plt.title(title)
    plt.imshow(img, cmap=color)
    plt.show()
    print("DEBUG: plot " + title)

# Store image
def store_image(filename, filepath, img):
    '''
    stores image as file name to file path
    '''
    # create output file path
    output_filepath = os.path.join(filepath, filename)
    
    # make sure the directory exists before writing to file
    os.makedirs(os.path.dirname(output_filepath), exist_ok=True)
    
    # write to file
    cv2.imwrite(output_filepath, img)
    print("DEBUG: set " + output_filepath)
```

### Step 3: Use color transform, gradients, etc., to create a thresholded binary image

```python
# test image filepaths
test_image_path = "../test_images/*.jpg"

# Create threshold binary image
def create_color_binary(img,
                          s_thresh = (170, 255),
                          sx_thresh = (20, 100)):
    '''
    returns final image 'color_binary' is a combination of binary
    thresholding the S channel (HLS) and binary thresholding the
    result of applying the Sobel operator in the x direction
    on the original image
    '''
    # Convert to HLS color space and separate the V channel
    hls = cv2.cvtColor(img, cv2.COLOR_RBG2HLS)
    l_channel = hls[:,:,1]
    s_channel = hls[:,:,2]
    
    # Sobel x
    sobelx = cv2.Sobel(l_channel, cv2.CV_64F, 1, 0) # Take the
                                                    # derivative in x
    abs_sobelx = np.absolute(sobelx) # Absolute x derivative to
                                     # accentuate lines away from
                                     # horizontal
    scaled_sobel = np.uint8(255*abs_sobelx / np.max(abs_sobelx))
    
    # Threshold x gradient
    sxbinary = np.zeros_like(scaled_sobel)
    sxbinary[(scaled_sobel >= sx_thresh[0]) & 
             (scaled_sobel <= sx_thresh[1])] = 1
    
    # Threshold color channel
    s_binary = np.zeros_like(s_channel)
    s_binary[(s_channel >= s_thresh[0]) &
             (s_channel <= s_thresh[1])] = 1
    
    # Stack each channel
    color_binary = np.dstack((np.zeros_like(sxbinary),
                              sxbinary,s_binary)) * 255
    return color_binary

# Filter lane
def lane_filter(fname):
    img = cv2.imread(fname)
    
    color_binary = create_color_binary(img)
    
    masked_lane = np.logical_and()
    
    

# Iterate over test images to create threshold binary
for fname in glob.glob(h_inpaths["test_images"]):
    img = cv2.imread(fname)
    
    color_binary = create_color_binary(img)
    
    # Store image
    store_image(h_filenames["thres_bin"]+os.path.basename(fname),
                h_outpaths["thres_bin"], color_binary)
    # Plot image
    plot_image(color_binary, "Step 3: Threshold Binary " + os.path.basename(fname))    
```

### Step 4: Apply a perspective transform

```python
# Apply perspective transform
def perspective_transform(img):
    '''
    returns undistort and perspective transformed image
    '''
    # 1) Undistort using mtx and dist
    undist = cv2.undistort(img, mtx, dist, None, mtx)
    # 2) get image shape
    img_size = (img.shape[1], img.shape[0])
    # TODO: replace hard-code source / destination points
    src = np.float32(
            [[200, img.shape[0]],
            [img.shape[1]-200, img.shape[0]],
            [(img.shape[1]/2)-40, (img.shape[0]/2)+90],
            [(img.shape[1]/2)+40, (img.shape[0]/2)+90]])
    dst = np.float32(
            [[300, img.shape[0]],
            [img.shape[1]-300, img.shape[0]],
            [300, 0],
            [img.shape[1]-300, 0]])
    # 3) get transform matrix
    M = cv2.getPerspectiveTransform(src, dst)
    # 4) warp image to top-down view
    warped = cv2.warpPerspective(undist, M, img_size)
    
    return warped

# Iterate over test images to apply perspective transform
for fname in glob.glob(h_inpaths["test_images"]):
    img = cv2.imread(fname)
    
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    warped = perspective_transform(rgb_img)

    # Store image
    store_image(h_filenames["pers_trans"]+os.path.basename(fname),
                h_outpaths["pers_trans"], warped)
    # Plot image
    plot_image(warped, "Step 4: Perspective Transform " + os.path.basename(fname)) 
```

### Step 5: Detect lane pixels and fit to find the lane boundary

```python
# detect lane pixels
def find_lane_pixels(binary_warped):
    '''
    return left and right pixels of detected lanes and 
    the output image
    '''
    # Take a histogram of the bottom half of the image
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:],
                       axis=0)
    # Create an output image to draw on and visualize the result
    out_img = np.dstack((binary_warped,binary_warped,binary_warped))
    
    # plot_image(out_img, "DEBUG") 
    print(out_img)
    
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    
    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50
    
    # Set heigh of windows - based on nwindows above and image shape
    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base
    
    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []
    
    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        
        # Draw the windows on the visualization image
        cv2.rectangle(out_img,
                     (win_xleft_low,win_y_low),
                     (win_xleft_high,win_y_high),
                      (0,255,0),
                      2) 
        cv2.rectangle(out_img, 
                    (win_xleft_low, win_y_low),
                    (win_xright_high, win_y_high),
                    (0,255,0), 2)
        
        # Identify the nonzero pixels i nx and y within the window #
        good_left_inds = ((nonzeroy >= win_y_low) & 
                         (nonzeroy < win_y_high) &
                         (nonzerox >= win_xleft_low) &
                         (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) &
                          (nonzeroy < win_y_high) &
                          (nonzerox >= win_xright_low) &
                          (nonzerox < win_xright_high)).nonzero()[0]
        
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        # If you found > minpix pixels, recenter next window
        # on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzero[good_left_inds]))
        if len(good_right_inds) > minpix:
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))
        
    # Concatenate the arrays of indices (previously was a list of
    # lists of pixels)
    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds]
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    
    return leftx, lefty, rightx, righty, out_img


# fits the lane boundary
def fit_polynomial(binary_warped):
    '''
    returns the image with visualization of left and right lane
    '''
    # Find lane pixels first
    leftx,lefty,rightx,righty,out_img = find_lane_pixels(binary_warped)
    
    # Fit a second order polynomial to each using `np.polyfit`
    left_fit = np.polyfit(lefty, leftx, 2)
    right_fit = np.polyfit(righty, rightx, 2)
    
    # Generate x and y values for plotting
    ploty = np.linspace(0,
                        binary_warped.shape[0]-1,
                        binary_warped.shape[0])
    try:
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty \
            + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty \
            + right_fit[2]
    except TypeError:
        # Avoids an error if `left` and `right_fit` are still non
        # or incorrect
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty
    
    # Visualization
    # Colors in the left and right lane regions
    out_img[lefty, leftx] = [255, 0, 0]
    out_img[righty, rightx] = [0, 0, 255]
    
    # Plots the left and right polynomials on the lane lines
    plt.plot(left_fitx, ploty, color='yellow')
    plt.plot(right_fitx, ploty, color='yellow')
    
    return out_img

# Iterate over test images to detect and apply lane boundary
for fname in glob.glob(h_inpaths["warped_images"]):
    img = cv2.imread(fname)
    print(fname)
    
    print(img.shape)
    
    plot_image(img, "DEBUG")
    
    # convert to grayscale
    # TODO: change this from gray to masked?
    # TODO: reason... cv2 format picture cannot have
    # TOOD: color channle
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    print(gray.shape)
    
    plot_image(gray, "DEBUG2")
    
    out_img = fit_polynomial(gray)
    
    print("done.")
    
    # Store image
    # store_image(h_filenames["fit_lane"]+os.path.basename(fname),
                # h_outpaths["fit_lane"], warped)
    # Plot image
    # plot_image(out_img, "Step 5: Fit lane boundary " + os.path.basename(fname)) 
```
